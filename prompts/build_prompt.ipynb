# Generador de Prompts para Codex

Este notebook genera prompts listos para usar con Codex (o cualquier asistente de código),
combinando tres archivos:

| Archivo | Rol |
|---------|-----|
| `AGENTS.md` | Contexto del proyecto: objetivos, pipeline, restricciones |
| `TASKS/*.md` | La tarea concreta que debe resolver el modelo |
| `prompts/codex_templates/*.txt` | Plantilla que une ambos en el formato correcto |

El resultado se guarda automáticamente en `runs/<timestamp>_prompt.txt`
y también en `last_prompt.txt` para acceso rápido.


## Celda 1 — Imports

```python
from pathlib import Path
from datetime import datetime
print("imports cargados")
```

## Celda 2 — Funciones base

- `load_file` lee cualquier archivo de texto.
- `summarize_agents` extrae solo las secciones relevantes de `AGENTS.md`
  para no saturar el prompt con contenido innecesario.
- `build_prompt` une todo, rellena la plantilla y guarda el resultado.


```python
from prompt import load_file, summarize_agents, build_prompt

print("Funciones cargadas correctamente.")
```

## Celda 3 — Configuracion

**Edita esta celda** para apuntar a tus archivos antes de continuar.


```python
AGENTS_PATH   = "../AGENTS.md"
TASK_PATH     = "../TASKS/task_split_dataset.md"
TEMPLATE_PATH = "python_feature.txt"

print(f"AGENTS   : {AGENTS_PATH}")
print(f"TASK     : {TASK_PATH}")
print(f"TEMPLATE : {TEMPLATE_PATH}")

# Verificar que los archivos existen antes de continuar
missing = [p for p in [AGENTS_PATH, TASK_PATH, TEMPLATE_PATH]
           if not Path(p).exists()]
if missing:
    print()
    print("ADVERTENCIA - archivos no encontrados:")
    for p in missing:
        print(f"  {p}")
else:
    print()
    print("Todos los archivos encontrados. Listo para generar.")
```

## Celda 4 — Generar el prompt

```python
prompt = build_prompt(
    agents_path   = AGENTS_PATH,
    task_path     = TASK_PATH,
    template_path = TEMPLATE_PATH,
)
```

## Celda 5 — Vista previa del prompt

Se muestran los primeros 2000 caracteres para no saturar la pantalla.


```python
print("=" * 50)
print("PROMPT GENERADO (primeros 2000 caracteres)")
print("=" * 50)
print()
print(prompt[:2000])

if len(prompt) > 2000:
    print()
    print(f"... [{len(prompt) - 2000} caracteres más — ver last_prompt.txt para el texto completo]")
```

## Celda 6 — Exportar prompt completo

Guarda el prompt en `last_prompt.txt` para copiarlo fácilmente
sin tener que buscar en la carpeta `runs/`.


```python
output_file = "last_prompt.txt"
Path(output_file).write_text(prompt, encoding="utf-8")
print(f"Prompt completo guardado en: {output_file}")
print(f"Longitud total: {len(prompt)} caracteres")
```
